```python
g = []
donot = ['Problem Statement Details','Description','Problem Statement Type','Created At','Youtube Link','Updated At']
for i in y:
    if len(i)>2 and i.strip() not in donot:
        g.append(i.strip())
    try:
        if int(i)>=0:
            g.append(i.strip())
    except:
        continue
```

```python
g
```

```python
all = []

for i in range(1,52):
    rurl = url + str(i)
    r = requests.get(rurl)
    data = r.text
    soup = bs4.BeautifulSoup(data)
    table = soup.find_all("tbody")[0]
    rows = table.find_all("tr")
    for x in rows:
        x= x.text
        y = x.split('\n')
        g = []
        for i in y:
            if len(i)>2 :
                g.append(i.strip())
            try:
                if int(i)>=0:
                    g.append(i.strip())
            except:
                continue
        all.append(g)
```

```python
all
```

```python
final = []
for i in all:
    if len(i)>2:
        final.append(i)
```

```python
final
```

```python
final[36]
```

```python
notfinal = []
cout = 0
for i in final:
    t = []
    try:
        desc = i.index('Description')
        pst = i.index('Problem Statement Type')
        t.extend(i[0:desc+1])
        t.append(''.join(i[desc+1:pst]))
        t.extend(i[pst:])
        notfinal.append(t)
    except:
        if len(i) > 10:
            print(len(i),final.index(i))
            cout +=1 
        continue
```

```python
for i in notfinal:
    print(len(i),notfinal.index(i))
```
